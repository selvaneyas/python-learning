```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import Lasso, Ridge
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
```

```python
# Generate synthetic dataset
X, y = make_regression(n_samples=100, n_features=10, noise=0.1, random_state=42)
```

```python
# Lasso Regression for Feature Selection
lasso = Lasso(alpha=0.1)
lasso.fit(X, y)
lasso_importance = np.abs(lasso.coef_)
```

```python
lasso
```

```python
lasso_importance
```

```python
# Ridge Regression for Feature Selection
ridge = Ridge(alpha=1.0)
ridge.fit(X, y)
ridge_importance = np.abs(ridge.coef_)
```

```python
ridge
```

```python
ridge_importance
```

```python
# Recursive Feature Elimination (RFE)
estimator = RandomForestRegressor(n_estimators=100, random_state=42)
rfe = RFE(estimator, n_features_to_select=5)
rfe.fit(X, y)
rfe_ranking = rfe.ranking_
```

```python
rfe
```

```python
rfe_ranking
```

```python
# Visualization
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Lasso Plot
ax[0].bar(range(len(lasso_importance)), lasso_importance, color='blue')
ax[0].set_title("Lasso Feature Importance")

# Ridge Plot
ax[1].bar(range(len(ridge_importance)), ridge_importance, color='red')
ax[1].set_title("Ridge Feature Importance")

# RFE Plot
ax[2].bar(range(len(rfe_ranking)), rfe_ranking, color='green')
ax[2].set_title("RFE Feature Ranking (Lower is better)")

plt.show()
```
