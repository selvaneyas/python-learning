# 1Ô∏è‚É£ Summary Statistics

```python
print(df.describe()) # Summary of numerical features
```

```python
print(df.describe(include='all')) # Summary of all features
```

```python
print(df.info()) # Data type information
```

# 2Ô∏è‚É£ Handling Missing Values

```python
print("Missing Values : \n", df.isnull().sum()) # Missing Values
```

```python
df.fillna(df.median(numeric_only= True), inplace=True) # Fill missing values with median
```

# 3Ô∏è‚É£ Data Visualization

```python
plt.figure(figsize=(10, 5))
```

## 1. Histogram

```python
# Histogram

df["age"].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

## 2. Boxplot

```python
# Boxplot for outlier detection

sns.boxplot(x = df["fare"])
plt.title("Fare Boxplot")
plt.show()
```

## 3. Heatmap

```python
# Correlation Heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

The **Titanic dataset** contains passenger details from the Titanic ship disaster. Below is a **summary of each column and its meaning**:

### **Titanic Dataset Column Summary**  
1. **survived** (*int64*)  
   - 0 ‚Üí Passenger did not survive  
   - 1 ‚Üí Passenger survived  

2. **pclass** (*int64*)  
   - 1 ‚Üí First class  
   - 2 ‚Üí Second class  
   - 3 ‚Üí Third class  

3. **sex** (*category*)  
   - male ‚Üí Male passenger  
   - female ‚Üí Female passenger  

4. **age** (*float64*)  
   - Age of the passenger (in years)  
   - Some values may be **NaN** (missing)  

5. **sibsp** (*int64*)  
   - Number of **siblings or spouses** aboard the Titanic  
   - Includes brothers, sisters, wives, and husbands  

6. **parch** (*int64*)  
   - Number of **parents or children** aboard  
   - Includes mothers, fathers, sons, and daughters  

7. **fare** (*float64*)  
   - Ticket price paid by the passenger  

8. **embarked** (*category*)  
   - Port where the passenger **boarded the Titanic**  
   - C ‚Üí Cherbourg  
   - Q ‚Üí Queenstown  
   - S ‚Üí Southampton  

9. **class** (*category*)  
   - Passenger class (same as `pclass`, but as a categorical variable)  

10. **who** (*category*)  
    - male ‚Üí Adult male  
    - female ‚Üí Adult female  
    - child ‚Üí Child passenger  

11. **adult_male** (*bool*)  
    - True ‚Üí Adult male  
    - False ‚Üí Not an adult male  

12. **deck** (*category*)  
    - Deck location where the passenger's cabin was  
    - Values: A, B, C, D, E, F, G  
    - Some values may be **NaN**  

13. **embark_town** (*category*)  
    - Full name of the **embarked** column:  
    - Cherbourg, Queenstown, Southampton  

14. **alive** (*category*)  
    - no ‚Üí Passenger did not survive  
    - yes ‚Üí Passenger survived  

15. **alone** (*bool*)  
    - True ‚Üí Passenger traveled alone (no family aboard)  
    - False ‚Üí Passenger had family aboard  

### **Key Notes**
- The dataset contains **missing values** in some columns (`age`, `deck`, `embark_town`).  
- `pclass` and `class` are the same information but in different formats (`int` vs `category`).  
- `sibsp` and `parch` can be combined to calculate total family members.
