Clustering is a fundamental unsupervised machine learning technique used to group similar data points together. 



## ùóßùòÑùóº ùóΩùóºùóΩùòÇùóπùóÆùóø ùó∞ùóπùòÇùòÄùòÅùó≤ùóøùó∂ùóªùó¥ ùóÆùóπùó¥ùóºùóøùó∂ùòÅùóµùó∫ùòÄ:

 üîπ Hierarchical Clustering ‚Äì A tree-based clustering method that creates a hierarchy of clusters.

 ‚úÖ Identifies hidden patterns in data.

 ‚úÖ Used in market research, bioinformatics, and recommendation systems.



## üìä ùó©ùó∂ùòÄùòÇùóÆùóπùó∂ùòáùó∂ùóªùó¥ ùó∏-ùó†ùó≤ùóÆùóªùòÄ & ùóõùó∂ùó≤ùóøùóÆùóøùó∞ùóµùó∂ùó∞ùóÆùóπ ùóñùóπùòÇùòÄùòÅùó≤ùóøùó∂ùóªùó¥

 üîπ k-Means: Assigns data points to clusters based on proximity to centroids.

 üîπ Dendrogram (Hierarchical Clustering): Illustrates how clusters merge at different levels.



## üöÄ ùóûùó≤ùòÜ ùóßùóÆùó∏ùó≤ùóÆùòÑùóÆùòÜùòÄ:

 üîπ k-Means is efficient but requires pre-defining the number of clusters.

 üîπ Hierarchical Clustering does not require specifying k but is computationally expensive for large datasets.

 üîπ Both methods are widely used for data exploration and pattern recognition.


## Import Required Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.datasets import make_blobs
from scipy.cluster.hierarchy import dendrogram, linkage
```

## Generate Sample Data for Clustering

```python
# Create a dataset with 3 clusters
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)
```

## k-Means Clustering Visualization

```python
# Apply k-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels_kmeans = kmeans.fit_predict(X)
```

```python
# Plot k-Means Clustering Result
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels_kmeans, palette="viridis", s=100, edgecolor="k")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label="Centroids")
plt.title("k-Means Clustering")
plt.legend()
plt.show()
```

## Hierarchical Clustering (Dendrogram) Visualization

```python
# Generate linkage matrix
linkage_matrix = linkage(X, method="ward")
```

```python
# Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

## Hierarchical Clustering (Clusters) Visualization

```python
# Apply Agglomerative Clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
labels_hierarchical = hierarchical.fit_predict(X)
```

```python
# Plot Hierarchical Clustering Result
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels_hierarchical, palette="coolwarm", s=100, edgecolor="k")
plt.title("Hierarchical Clustering")
plt.show()
```

## Python Code for Clustering Visualization

```python
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns  
from sklearn.cluster import KMeans, AgglomerativeClustering  
from sklearn.datasets import make_blobs  
from scipy.cluster.hierarchy import dendrogram, linkage  
```

```python
# Generate Sample Data  
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)  
```

```python
# k-Means Clustering  
kmeans = KMeans(n_clusters=3, random_state=42)  
labels_kmeans = kmeans.fit_predict(X)
```

```python
# k-Means Visualization  
plt.figure(figsize=(8, 6))  
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels_kmeans, palette="viridis", s=100, edgecolor="k")  
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label="Centroids")  
plt.title("k-Means Clustering")  
plt.legend()  
plt.show()
```

```python
# Hierarchical Clustering - Dendrogram  
linkage_matrix = linkage(X, method="ward")  
plt.figure(figsize=(10, 6))  
dendrogram(linkage_matrix)  
plt.title("Hierarchical Clustering Dendrogram")  
plt.xlabel("Data Points")  
plt.ylabel("Distance")  
plt.show()
```
